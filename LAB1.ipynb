```python
import pandas as pd

# Đường dẫn đến file CSV
file_path = r'd:\Hsau\dulieuxettuyendaihoc.csv'

try:
    # Đọc dữ liệu từ file
    data = pd.read_csv(file_path, encoding='utf-8')

    # Thống kê dữ liệu thiếu trong cột T1
    missing_count_before = data['T1'].isnull().sum()
    print(f"Số lượng dữ liệu thiếu trước khi hiệu chỉnh: {missing_count_before}")

    # Tính giá trị trung bình (Mean) của cột T1, bỏ qua các giá trị NaN
    mean_value = data['T1'].mean()
    print(f"Giá trị trung bình của cột 'T1': {mean_value}")

    # Thay thế dữ liệu thiếu bằng giá trị trung bình
    data['T1'].fillna(mean_value, inplace=True)

    # Thống kê lại dữ liệu thiếu sau khi thay thế
    missing_count_after = data['T1'].isnull().sum()
    print(f"Số lượng dữ liệu thiếu sau khi hiệu chỉnh: {missing_count_after}")

    # Lưu dữ liệu đã hiệu chỉnh vào file mới
    output_file = r'd:\Hsau\dulieuxettuyendaihoc_cleaned_mean.csv'
    data.to_csv(output_file, index=False, encoding='utf-8')
    print(f"Dữ liệu đã hiệu chỉnh được lưu tại: {output_file}")

except FileNotFoundError:
    print(f"File '{file_path}' không tồn tại. Vui lòng kiểm tra lại.")
except KeyError:
    print("Cột 'T1' không tồn tại trong dữ liệu. Vui lòng kiểm tra lại tên cột.")
except Exception as e:
    print(f"Đã xảy ra lỗi: {e}")
```

```python
import pandas as pd

# Đường dẫn đến file CSV
file_path = r'd:\Hsau\dulieuxettuyendaihoc.csv'

try:
    # Đọc dữ liệu từ file
    data = pd.read_csv(file_path, encoding='utf-8')

    # Tổng số dòng
    total_rows = len(data)

    # Tần số dữ liệu thiếu (số giá trị NaN) cho từng cột
    missing_count = data.isnull().sum()

    # Tần suất thiếu (%) cho từng cột
    missing_percentage = (missing_count / total_rows) * 100

    # Tạo bảng tổng hợp
    missing_table = pd.DataFrame({
        'Cột': missing_count.index,
        'Tần số thiếu': missing_count.values,
        'Tần suất thiếu (%)': missing_percentage.values
    })

    # Lọc chỉ các cột có dữ liệu thiếu
    missing_table = missing_table[missing_table['Tần số thiếu'] > 0]

    print("Bảng tần số và tần suất dữ liệu thiếu:")
    print(missing_table)

except FileNotFoundError:
    print(f"File '{file_path}' không tồn tại. Vui lòng kiểm tra lại.")
except Exception as e:
    print(f"Đã xảy ra lỗi: {e}")
```

```python
import pandas as pd

# Đường dẫn đến file CSV
file_path = r'd:\Hsau\dulieuxettuyendaihoc.csv'

try:
    # Đọc dữ liệu từ file
    data = pd.read_csv(file_path, encoding='utf-8')

    # Kiểm tra và thay thế dữ liệu thiếu bằng giá trị trung bình
    for column in data.columns:
        if data[column].isnull().sum() > 0:  # Kiểm tra nếu cột có dữ liệu thiếu
            if data[column].dtype in ['float64', 'int64']:  # Chỉ áp dụng cho cột số
                mean_value = data[column].mean()  # Tính giá trị trung bình
                print(f"Thay thế dữ liệu thiếu trong cột '{column}' bằng giá trị trung bình: {mean_value}")
                data[column].fillna(mean_value, inplace=True)
            else:
                print(f"Cột '{column}' không phải kiểu số, bỏ qua.")

    # Lưu dữ liệu đã hiệu chỉnh vào file mới
    output_file = r'd:\Hsau\dulieuxettuyendaihoc_mean_cleaned.csv'
    data.to_csv(output_file, index=False, encoding='utf-8')
    print(f"Dữ liệu đã hiệu chỉnh được lưu tại: {output_file}")

except FileNotFoundError:
    print(f"File '{file_path}' không tồn tại. Vui lòng kiểm tra lại.")
except Exception as e:
    print(f"Đã xảy ra lỗi: {e}")
```

```python
import pandas as pd

# Đường dẫn đến file CSV
file_path = r'd:\Hsau\dulieuxettuyendaihoc.csv'

try:
    # Đọc dữ liệu từ file
    data = pd.read_csv(file_path, encoding='utf-8')

    # Lấy danh sách các cột liên quan đến điểm số (giả sử tên các cột có từ "T" hoặc là cột số)
    score_columns = [col for col in data.columns if data[col].dtype in ['float64', 'int64']]

    print("Các cột điểm số cần xử lý:")
    print(score_columns)

    # Xử lý dữ liệu thiếu cho từng cột
    for column in score_columns:
        if data[column].isnull().sum() > 0:  # Chỉ xử lý nếu có dữ liệu thiếu
            mean_value = data[column].mean()  # Tính giá trị trung bình
            print(f"Thay thế dữ liệu thiếu trong cột '{column}' bằng giá trị trung bình: {mean_value}")
            data[column].fillna(mean_value, inplace=True)

    # Kiểm tra xem còn dữ liệu thiếu không
    remaining_missing = data.isnull().sum().sum()
    if remaining_missing == 0:
        print("\nTất cả dữ liệu thiếu đã được xử lý.")
    else:
        print(f"\nVẫn còn {remaining_missing} giá trị thiếu trong dữ liệu.")

    # Lưu dữ liệu đã hiệu chỉnh vào file mới
    output_file = r'd:\Hsau\dulieuxettuyendaihoc_all_cleaned.csv'
    data.to_csv(output_file, index=False, encoding='utf-8')
    print(f"Dữ liệu đã hiệu chỉnh được lưu tại: {output_file}")

except FileNotFoundError:
    print(f"File '{file_path}' không tồn tại. Vui lòng kiểm tra lại.")
except Exception as e:
    print(f"Đã xảy ra lỗi: {e}")
```

```python
import pandas as pd

# Đường dẫn đến file CSV
file_path = r'd:\Hsau\dulieuxettuyendaihoc.csv'

try:
    # Đọc dữ liệu từ file
    data = pd.read_csv(file_path, encoding='utf-8')

    # Tạo các biến trung bình môn cho từng năm học
    for year in [1, 2, 3]:  # 1: lớp 10, 2: lớp 11, 3: lớp 12
        tbm_column = f'TBM{year}'
        t, l, h, s, v, x, d, n = [f'{subject}{year}' for subject in ['T', 'L', 'H', 'S', 'V', 'X', 'D', 'N']]

        # Kiểm tra nếu các cột cần thiết có tồn tại
        required_columns = [t, l, h, s, v, x, d, n]
        if all(col in data.columns for col in required_columns):
            # Tính TBM theo công thức
            data[tbm_column] = (data[t] * 2 + data[l] + data[h] + data[s] +
                                data[v] * 2 + data[x] + data[d] + data[n]) / 10
            print(f"Tạo cột {tbm_column} thành công.")
        else:
            print(f"Các cột cần thiết để tính {tbm_column} không đầy đủ. Bỏ qua.")

    # Lưu dữ liệu đã hiệu chỉnh vào file mới
    output_file = r'd:\Học tập\Năm 3\Học sâu\Thực hành\dulieuxettuyendaihoc_with_TBM.csv'
    data.to_csv(output_file, index=False, encoding='utf-8')
    print("\nDữ liệu đã được lưu tại:", output_file)

    # Hiển thị 5 dòng đầu tiên với các cột mới
    print("\nDữ liệu sau khi thêm các cột TBM:")
    print(data[[f'TBM{year}' for year in [1, 2, 3]]].head())

except FileNotFoundError:
    print(f"File '{file_path}' không tồn tại. Vui lòng kiểm tra lại.")
except Exception as e:
    print(f"Đã xảy ra lỗi: {e}") 
    
import pandas as pd

# Đường dẫn đến file CSV
file_path = r'd:\Hsau\dulieuxettuyendaihoc.csv'

try:
    # Đọc dữ liệu từ file
    data = pd.read_csv(file_path, encoding='utf-8')

    # Tạo các biến trung bình môn cho từng năm học
    for year in [1, 2, 3]:  # 1: lớp 10, 2: lớp 11, 3: lớp 12
        tbm_column = f'TBM{year}'
        t, l, h, s, v, x, d, n = [f'{subject}{year}' for subject in ['T', 'L', 'H', 'S', 'V', 'X', 'D', 'N']]

        # Kiểm tra nếu các cột cần thiết có tồn tại
        required_columns = [t, l, h, s, v, x, d, n]
        if all(col in data.columns for col in required_columns):
            # Tính TBM theo công thức
            data[tbm_column] = (data[t] * 2 + data[l] + data[h] + data[s] +
                                data[v] * 2 + data[x] + data[d] + data[n]) / 10
            print(f"Tạo cột {tbm_column} thành công.")
        else:
            print(f"Các cột cần thiết để tính {tbm_column} không đầy đủ. Bỏ qua.")

    # Lưu dữ liệu đã hiệu chỉnh vào file mới
    output_file = r'd:\Hsau\dulieuxettuyendaihoc_with_TBM.csv'
    data.to_csv(output_file, index=False, encoding='utf-8')
    print("\nDữ liệu đã được lưu tại:", output_file)

    # Hiển thị 5 dòng đầu tiên với các cột mới
    print("\nDữ liệu sau khi thêm các cột TBM:")
    print(data[[f'TBM{year}' for year in [1, 2, 3]]].head())

except FileNotFoundError:
    print(f"File '{file_path}' không tồn tại. Vui lòng kiểm tra lại.")
except Exception as e:
    print(f"Đã xảy ra lỗi: {e}")
```
